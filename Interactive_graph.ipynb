```python
import math
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import plotly.offline as pyo
import plotly.express as px
import plotly.graph_objs as go
```

```python
 df = pd.read_csv("fc1.csv", index_col= "LoggedAt")
 df
```

```python
adf = df.groupby(["LocationName", "FaultName"]).size()
adf
   
```

```python
adf = adf.to_frame(name = 'FaultCount').reset_index()
#adf = adf.nlargest(300,"FaultCount") 
```

```python
adf = adf.sort_values("FaultName")
adf
```

```python
datax = adf["LocationName"].tolist()
datay = adf["FaultName"].tolist()
```

```python
FaultCount = adf["FaultCount"].tolist()
FaultCount = np.array(FaultCount)
#FaultCount = np.log(FaultCount)
```

```python
sizeref = 2. * max(FaultCount) / (80)
sizeref
```

```python
data = [go.Scatter(x = datax, 
        y = datay,
        text = adf['FaultCount'],
        mode = "markers", 
        marker=dict(
        size=FaultCount/max(FaultCount) *10000 ,
        #sizeref = 2. * max(FaultCount) / (48),
        color=FaultCount, #set color equal to a variable
        colorscale='Viridis',
        sizemode = 'area', 
        showscale=True))]
pyo.plot(data)
```
